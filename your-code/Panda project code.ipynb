# Importar las librerías pandas y numpy para este proyecto.

```python
import pandas as pd
import numpy as np
```

# Importar con ayuda de pandas los archivos .txt a trabajar.

```python
package = pd.read_csv('../package.txt', sep = '\t')
product = pd.read_csv('../product.txt', sep = '\t')
```

# Decidí empezar a trabajar con el dataframe package.

```python
package
```

#### Revisar el tipo de datos que tiene nuestra df package.

```python
package.dtypes
```

#### Cambiar los nombres de las columnas.

```python
col_PCKnames ={'PRODUCTID':'Product Id','PRODUCTNDC':'P NCD','NDCPACKAGECODE':'NCD PCK Code','PACKAGEDESCRIPTION':'PCK Description','STARTMARKETINGDATE':'Start Mkt Date','ENDMARKETINGDATE':'End Mkt Date','NDC_EXCLUDE_FLAG':'NCD Exclude Flag','SAMPLE_PACKAGE':'Sample PCK'}
```

```python
package = package.rename(columns = col_PCKnames)
```

#### Revisar las columnas con datos faltantes.

```python
null_package = package.isnull().sum()
null_package[null_package > 0]
```

##### Como la columna End Mkt Date tiene solamente 5K de los 250K, decidí eliminarla.

```python
drop_package = list(null_package[null_package > 200000].index)
package = package.drop(drop_package, axis=1)
```

#### Al revisar los tipos de datos la columna Start Mkt Date debe de estar en formato datetime.

Los pasos siguientes son para hacer este cambio.

```python
package['Start Mkt Date'] = package['Start Mkt Date'].astype(str)
```

```python
package['Start Mkt Date'] = package['Start Mkt Date'].str.replace('[.]0','')
```

```python
package['Start Mkt Date'] = pd.to_datetime(package['Start Mkt Date'], format='%Y%m%d')
```

#### Reemplacé los faltantes de la columna Start Mkt Date por 0.

```python
package['Start Mkt Date'].fillna(0, inplace=True)
```

```python
null_package = package.isnull().sum()
null_package[null_package > 0]
```

# Así quedó la base final.

```python
package
```

# Los pasos siguientes son para el dataframe product.

```python
product
```

#### Primero revisé el tipo de datos.

```python
product.dtypes
```

#### Generé un diccionario para cambiar los nombres de mis columnas.

```python
colP_ornames = product.columns
colP_names = ['Product Id','P NDC','Producity Name','Propietary Name','Propietary Name Suffix','Non Propietary Name','Dosage Form Name','Route Name','Start Mkt Date','End Mkt Date','Mkt Category Name','Application number','Labeler Name','Substance Name','Active Numerator Strength','Active Ingredient Unit','Phram Classes','Desachedule','NDC Exclude Flag','LRC Through']
dictP_names = dict(zip(colP_ornames, colP_names))
```

```python
product = product.rename(columns = dictP_names)
```

#### Revisé las columnas con datos faltantes. Al ser menores a la mitad, decidí dejar todas.

```python
null_product = product.isnull().sum()
null_product[null_product > 0]
```

#### Para hacer el cambio de tipo de dato de float a datatime, primero los pasé a str para eliminar el ".0", después cambié un dato erroneo en la columna "End Mkt Date" por NaN.

```python
product[['Start Mkt Date','End Mkt Date','LRC Through']] = product[['Start Mkt Date','End Mkt Date','LRC Through']].astype(str)
```

```python
product['Start Mkt Date']=product['Start Mkt Date'].str.replace('[.]0','')
product['End Mkt Date']=product['End Mkt Date'].str.replace('[.]0','')
product['LRC Through']=product['LRC Through'].str.replace('[.]0','')
```

```python
product['Start Mkt Date']= pd.to_datetime(product['Start Mkt Date'], format='%Y%m%d')
```

```python
product.loc[product['End Mkt Date'] == '30310209', 'End Mkt Date'] = np.nan
```

```python
product['End Mkt Date']= pd.to_datetime(product['End Mkt Date'], format='%Y%m%d')
```

```python
product['LRC Through']= pd.to_datetime(product['LRC Through'], format='%Y%m%d')
```

```python
product.dtypes
```

#### Reemplacé los valores NaN por un 0.

```python
product.fillna(0, inplace=True)
```

```python
null_product = product.isnull().sum()
null_product[null_product > 0]
```

# Así quedó la base después de todos los cambios.

```python
product
```

# Decidí hacer un merge con las dos bases, eligiendo algunas columnas del dataframe product.

```python
product_tomerge = product[['Product Id','Producity Name','Dosage Form Name','Mkt Category Name','Labeler Name']]
the_merge = pd.merge(package, product_tomerge, on='Product Id')
```
